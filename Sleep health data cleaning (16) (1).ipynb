## ◻ Average sleep duration of each gender

```python
avg_sleep_gender = data.groupby(['Gender'])['Sleep_Duration'].mean().reset_index()
```

```python
avg_sleep_gender
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Gender',y='Sleep_Duration',data=avg_sleep_gender)
plt.xlabel("Gender")
plt.ylabel("Average Sleep Duration")
plt.title("Average Sleep Duration of Each Gender")
ax.bar_label(ax.containers[0],fmt='%.1f');
```

### ◽ From this graph we can see that average Sleep Duration of Each Gender

## ◻ How does sleep duration vary across different age groups?

```python
average_sleep_duration = data.groupby("Age")["Sleep_Duration"].mean().reset_index()
```

```python
average_sleep_duration
```

```python
plt.plot(average_sleep_duration["Age"], average_sleep_duration["Sleep_Duration"], marker='o')
plt.xlabel("Age")
plt.ylabel("Sleep_Duration")
plt.title("Sleep Duration across Different Age Groups")
plt.grid(True)
plt.show()
```

### ◽ From this graph Sleep duration across different Age groups, we can see that the highest age groups got the maximum duration of sleep.
​

## ◻ Which occupation have more stress level

```python
occup_stress_max.sort_values(by='Stress_Level',ascending=False)
```

```python
occup_stress_max[occup_stress_max['Stress_Level']==8]
```

### Bar Plot

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x='Occupation',y='Stress_Level',data=occup_stress_max)
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0],fmt='%.1f');
```

### ◽ From this graph we can see that Occupations like Doctor, Nurse, Sales- Representative, Scientist,Software Engineer have the highest stress level.

##  ◻️ Bar Plot

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Quality_of_Sleep',y='Age',data=data)
plt.title("Bar Plot")
ax.bar_label(ax.containers[0],fmt='%.1f');
```

## ◻ Line Chart

```python
plt.figure(figsize=(10,6))
plt.xlabel('Quality_of_Sleep')
plt.ylabel('Stress_Level')
plt.title('Line charts')
sns.lineplot(data=data,x='Quality_of_Sleep', y='Stress_Level')
plt.grid(True)
plt.show()
```

### ◽ This is a graph of the dependence of the level of stress and the quality of sleep. It can be seen that the lower the stress level have  higher the quality of sleep.

## ◻ Bar plot

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Gender", y="Sleep_Duration", hue="Occupation", data=data, palette="Set2")
plt.title("Sleep Duration by Gender and Occupation")
plt.show()
```

## ◻ Pie chart for percentage of each class of sleep disorder
